Creating files of 200,400,600,800,1000 mb and converting their content into uppercase . Then note their time also.

```python
import os, random, string, time

def generate_file(filename, size_mb):
    """Creates a file with random text of specified size (in MB)."""
    size_bytes = size_mb * 1024 * 1024
    chunk = ''.join(random.choices(string.ascii_letters + ' ', k=1024*16))
    written = 0
    with open(filename, 'w') as f:
        while written < size_bytes:
            to_write = chunk[:min(len(chunk), size_bytes - written)]
            f.write(to_write)
            written += len(to_write)
    print(f"Generated {size_mb}MB file: {filename}")

# Create files
sizes = [200, 400, 600, 800, 1000]
for sz in sizes:
    generate_file(f"file{sz}.txt", sz)
```

```python
def convert_to_uppercase(input_file, output_file):
    """Reads a file and writes an uppercase version."""
    start_time = time.time()
    with open(input_file, 'r', encoding='utf-8', errors='ignore') as fin, \
         open(output_file, 'w', encoding='utf-8') as fout:
        for chunk in iter(lambda: fin.read(1 << 20), ''):
            fout.write(chunk.upper())
    elapsed = time.time() - start_time
    print(f"Processed {input_file} → {output_file} in {elapsed:.3f} seconds")

# Convert each file
for sz in sizes:
    convert_to_uppercase(f"file{sz}.txt", f"uppercase_{sz}.txt")
```
